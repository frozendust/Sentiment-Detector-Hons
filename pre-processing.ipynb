```python
import nltk
#imports the list of stop words.
from nltk.corpus import stopwords
#Used to tokenize the text into a sentace or just words.
from nltk.tokenize import sent_tokenize, word_tokenize
#imort regular exprasions
import re
#Import text blob for sentiment demo
from textblob import TextBlob

#create a list of stopp words
stopWords = set(stopwords.words('english')) 

#Creating the The original sentance
text= "  Backgammon is one of the oldest known board games. Its history can be traced back nearly 5,000 years to archeological discoveries in the Middle East. It is a two player game where each player has fifteen checkers which move between twenty-four points according https://regex101.com/library/SqcCO5 to the roll of two  dice."
```

```python
print("##############################################################################################################################")
print("###################################################Text Pre-Processing########################################################")
print("##############################################################################################################################")


print("")
print("The original Text")
print(text)

print("")
print("####################################################")
print("####################################################")
#Convert to lowercase
print("Convert all to lowercase")
text = text.lower()
print(text)

#filter out url
print("")
print("####################################################")
print("####################################################")
print("Removing url")
pattern = r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*'

text= re.sub(pattern, "", text)
print(text)

#filter out numbers
print("")
print("####################################################")
print("####################################################")
print("Removing numbers")
pattern = r"[\d]"
text= re.sub(pattern, "", text)
print(text)

#filter out commas and stops
print("")
print("####################################################")
print("####################################################")
print("Removing commasa and stops")
pattern = r"[^\w]"
text= re.sub(pattern, " ", text)
print(text)

print("")
print("####################################################")
print("####################################################")
print("Removing white space")
text = text.strip()
print(text)

#################################################
#Breaking up into sentances first then words
#Sentance = sent_tokenize(text)
#print(Sentance)

#making a list of words from the setancees text
#for sentance in Sentance:
#    words = word_tokenize(sentance)
#    print(words)
#################################################
print("")
print("####################################################")
print("####################################################")
print("Tokenizing all the words")
words = word_tokenize(text)
print(words)

#Filtering out stop words using  list comprehensions
print("")
print("####################################################")
print("####################################################")
print("Removing stop words")
wordsFiltered = [word for word in words if not word in stopWords]
#################################################
#removing stoping using loops
#wordsFiltered = [] #array to store the new text once the words are filtered out.
#loop to filter out words.
#for w in words:
#    if w not in stopWords:
#        wordsFiltered.append(w)

print("Here is the filter")
print(wordsFiltered)
print('')

print("##############################################################################################################################")
print("###################################################Example Sentiment Detection################################################")
print("##############################################################################################################################")

###################################################
#text blob sentiment
###################################################
#combines the list into one string for textblob
seperator = ' '
wordsText = seperator.join(wordsFiltered)
#creating a textblob object
blob = TextBlob(wordsText)
blob1 = TextBlob("Analytics Vidhya is a great platform to learn data science.")
#blob.sentiment()
blob1.tags
#blob1.sentiment()
#blob.sentences
print("")
print("Example sentiment text using TextBlob")
print("original:")
print(blob1)
print("")
print("Sentiment:")
for sentence in blob1.sentences:
    print(sentence.sentiment)

# pre-processed string sentiemnt
print("")
print("pre-processed text:")
print(wordsText)
print("")
print("Our pre-processes test sentiment")
for sentence in blob.sentences:
    print(sentence.sentiment)
```
